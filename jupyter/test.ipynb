```python
import json                                                                                          
import glob                                                                                          
from pathlib import Path                                                                             
from collections import Counter                                                                      
import pandas as pd                                                                                  
                                                                                                    
# 데이터 경로                                                                                        
DATA_ROOT = Path("dataset/Mono_Camera_Semantic_Segmentation")                                        
LABELS_DIR = DATA_ROOT / "labels"                                                                    
                                                                                                    
# 클래스 카운트                                                                                      
class_counts = Counter()                                                                             
                                                                                                    
# JSON 파일 순회                                                                                     
label_files = glob.glob(str(LABELS_DIR / "*.json"))                                                  
print(f"Total label files: {len(label_files)}")                                                      
                                                                                                    
for label_path in label_files:                                                                       
    with open(label_path, 'r', encoding='utf-8') as f:                                               
        annotation = json.load(f)                                                                    
                                                                                                    
    for obj in annotation.get("objects", []):                                                        
        if obj.get("deleted", 0) == 0:                                                               
            label = obj.get("label", "unlabeled")                                                    
            class_counts[label] += 1                                                                 
                                                                                                                                                                
class_df = pd.DataFrame(                                                                             
    class_counts.most_common(),                                                                      
    columns=['class_name', 'count']                                                                  
)                                                                                                    
class_df['class_id'] = range(len(class_df))                                                          
class_df = class_df[['class_id', 'class_name', 'count']].reset_index(drop=True)
class_df
```
